```python
#using case
query="""
select Age,
case
when Age<30 then 'Young'
when Age Between 30 and 50 then 'Adult'
else 'Senior'
end as Age_group

from telecom_churn_customer1;
"""
pd.read_sql_query(query,conn)     
```

```python
#subquery
query="""
select *
from telecom_churn_customer1
where "Total Revenue">
(
select avg("Total Revenue")
from telecom_churn_customer1
);
"""
pd.read_sql_query(query,conn)     
```
